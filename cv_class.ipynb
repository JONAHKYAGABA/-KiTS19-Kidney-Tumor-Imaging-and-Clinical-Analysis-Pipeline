```python
# -*- coding: utf-8 -*-
"""cv_class.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1YgcR54_eIVMafqAbuLgwf_0DG2ONBP9E
"""
# ### Step 2: Resample Images
import SimpleITK as sitk
from tqdm import tqdm
import os

def resample_sitk(image, new_spacing=[1.5, 1.5, 1.5], is_label=False):
    """Resample a SimpleITK image to a new spacing."""
    original_spacing = image.GetSpacing()
    original_size = image.GetSize()
    new_size = [int(round(osz * ospc / nspc)) for osz, ospc, nspc in zip(original_size, original_spacing, new_spacing)]

    resampler = sitk.ResampleImageFilter()
    resampler.SetSize(new_size)
    resampler.SetOutputSpacing(new_spacing)
    resampler.SetOutputOrigin(image.GetOrigin())
    resampler.SetOutputDirection(image.GetDirection())
    resampler.SetInterpolator(sitk.sitkNearestNeighbor if is_label else sitk.sitkLinear)
    return resampler.Execute(image)

# Define paths for original and resampled data
data_dir = '/content/dataset/kits19-master/data'
resampled_dir = '/content/drive/MyDrive/kits19_resampled'
os.makedirs(resampled_dir, exist_ok=True)

# Process training cases (0 to 209)
for case in tqdm(range(210)):
    case_str = f'case_{case:05d}'
    imaging_path = os.path.join(data_dir, case_str, 'imaging.nii.gz')
    segmentation_path = os.path.join(data_dir, case_str, 'segmentation.nii.gz')

    # Load images with SimpleITK
    imaging_sitk = sitk.ReadImage(imaging_path)
    segmentation_sitk = sitk.ReadImage(segmentation_path, sitk.sitkUInt8)  # Ensure label image is uint8

    # Resample images
    resampled_imaging = resample_sitk(imaging_sitk)
    resampled_segmentation = resample_sitk(segmentation_sitk, is_label=True)

    # Save resampled files
    case_dir = os.path.join(resampled_dir, case_str)
    os.makedirs(case_dir, exist_ok=True)
    sitk.WriteImage(resampled_imaging, os.path.join(case_dir, 'imaging_resampled.nii.gz'))
    sitk.WriteImage(resampled_segmentation, os.path.join(case_dir, 'segmentation_resampled.nii.gz'))


# ### Step 3: Preprocess Metadata
import json
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load metadata from kits.json
with open('/content/dataset/kits19-master/data/kits.json', 'r') as f:
    kits_data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(kits_data)

# Handle missing values
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Encode categorical variables (example: histology if present)
if 'histology' in df.columns:
    df = pd.get_dummies(df, columns=['histology'], prefix='histology')

# Normalize continuous variables
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Save preprocessed metadata as a pickle file
df.to_pickle('/content/drive/MyDrive/kits19_metadata_preprocessed.pkl')

# ### Step 4: Exploratory Data Analysis (EDA)
import seaborn as sns
import matplotlib.pyplot as plt

# Load original metadata for EDA
with open('/content/dataset/kits19-master/data/kits.json', 'r') as f:
    kits_data_original = json.load(f)

df_original = pd.DataFrame(kits_data_original)
df_original.set_index('case_id', inplace=True)

# Display summary statistics
print("Summary Statistics:")
print(df_original.describe())

# Histogram of age at nephrectomy
plt.figure(figsize=(8, 6))
sns.histplot(df_original['age_at_nephrectomy'], bins=20)
plt.title('Distribution of Age at Nephrectomy')
plt.xlabel('Age')
plt.show()

# Count plot of tumor histologic subtype
plt.figure(figsize=(10, 6))
sns.countplot(y='tumor_histologic_subtype', data=df_original)
plt.title('Distribution of Tumor Histologic Subtype')
plt.show()

# Box plot of radiographic size by malignancy
plt.figure(figsize=(8, 6))
sns.boxplot(x='malignant', y='radiographic_size', data=df_original)
plt.title('Radiographic Size by Malignancy')
plt.show()

# ### Step 5: Visualize Sample Cases with Metadata
import nibabel as nib
import numpy as np

def load_resampled_case(case_id):
    """Load resampled imaging and segmentation data for a given case."""
    case_str = f'case_{case_id:05d}'
    resampled_dir = '/content/drive/MyDrive/kits19_resampled'
    imaging_path = os.path.join(resampled_dir, case_str, 'imaging_resampled.nii.gz')
    segmentation_path = os.path.join(resampled_dir, case_str, 'segmentation_resampled.nii.gz')

    if not os.path.exists(imaging_path) or not os.path.exists(segmentation_path):
        raise FileNotFoundError(f"Resampled files for {case_str} not found in {resampled_dir}")

    volume = nib.load(imaging_path)
    segmentation = nib.load(segmentation_path)
    return volume, segmentation

def visualize_case_with_metadata(case_id, df_metadata, output_dir):
    """Visualize the middle slice of a case with segmentation overlay and metadata."""
    # Load resampled data
    case_str = f'case_{case_id:05d}'
    volume, segmentation = load_resampled_case(case_id)
    volume_data = volume.get_fdata()
    segmentation_data = segmentation.get_fdata()

    # Select middle slice
    slice_idx = volume_data.shape[0] // 2

    # Create colored segmentation (kidney: red, tumor: blue, background: transparent)
    seg = segmentation_data[slice_idx]
    seg_colored = np.zeros((seg.shape[0], seg.shape[1], 3), dtype=np.uint8)
    seg_colored[seg == 1] = [255, 0, 0]  # Kidney in red
    seg_colored[seg == 2] = [0, 0, 255]  # Tumor in blue

    # Normalize volume slice to 0-255
    volume_slice = volume_data[slice_idx]
    hu_min, hu_max = -512, 512  # Typical HU range for CT scans
    volume_slice = np.clip(volume_slice, hu_min, hu_max)
    volume_slice = (volume_slice - hu_min) / (hu_max - hu_min) * 255
    volume_rgb = np.stack([volume_slice]*3, axis=-1).astype(np.uint8)

    # Overlay segmentation on volume
    alpha = 0.5
    segbin = seg > 0  # Mask for non-background
    overlayed = np.where(
        segbin[..., None],
        np.round(alpha * seg_colored + (1 - alpha) * volume_rgb).astype(np.uint8),
        volume_rgb
    )

    # Get metadata for the case
    case_metadata = df_metadata.loc[case_str]
    key_fields = ['age_at_nephrectomy', 'gender', 'body_mass_index', 'malignant', 'tumor_histologic_subtype']
    metadata_text = '\n'.join([f"{field}: {case_metadata[field]}" for field in key_fields])

    # Create figure with image and metadata
    fig, ax = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios': [3, 1]})
    ax[0].imshow(overlayed)
    ax[0].set_title(f'Case {case_str} - Slice {slice_idx}')
    ax[0].axis('off')
    ax[1].text(0.5, 0.5, metadata_text, fontsize=12, ha='center', va='center')
    ax[1].axis('off')

    # Save the figure
    os.makedirs(output_dir, exist_ok=True)
    plt.savefig(os.path.join(output_dir, f'{case_str}_summary.png'))
    plt.close()

# Visualize a sample of cases (0 to 4) with metadata
selected_cases = [0, 1, 2, 3, 4]  # Sample of 5 cases
output_dir = '/content/drive/MyDrive/kits19_visualizations'

for case_id in selected_cases:
    visualize_case_with_metadata(case_id, df_original, output_dir)
print(f"Visualizations with metadata saved to {output_dir}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import nibabel as nib
import SimpleITK as sitk
from scipy import stats
from sklearn.preprocessing import StandardScaler
from lifelines import KaplanMeierFitter, CoxPHFitter
from lifelines.statistics import logrank_test
import warnings
warnings.filterwarnings('ignore')

# Set paths (update these as needed)
preprocessed_metadata_path = '/content/drive/MyDrive/kits19_metadata_preprocessed.pkl'
original_metadata_json = '/content/dataset/kits19-master/data/kits.json'
resampled_dir = '/content/drive/MyDrive/kits19_resampled'
output_dir = '/content/drive/MyDrive/kits19_analysis_results'

# Create output directory
os.makedirs(output_dir, exist_ok=True)

# Load preprocessed and original metadata
df_preprocessed = pd.read_pickle(preprocessed_metadata_path)

# Load original metadata for analysis
import json
with open(original_metadata_json, 'r') as f:
    kits_data_original = json.load(f)

df = pd.DataFrame(kits_data_original)
df.set_index('case_id', inplace=True)

# Convert nested dictionaries into flat columns
for col in df.columns:
    if isinstance(df[col].iloc[0], dict):
        nested_df = pd.json_normalize(df[col])
        nested_df.columns = [f"{col}_{subcol}" for subcol in nested_df.columns]
        df = pd.concat([df.drop(col, axis=1), nested_df], axis=1)

# Convert string values to appropriate types
for col in df.columns:
    if df[col].dtype == 'object':
        # Try to convert to numeric if applicable
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass

# Extract TNM stages into separate columns
if 'pathology_t_stage' in df.columns:
    df['t_stage_numeric'] = df['pathology_t_stage'].str.extract('(\d+)').astype(float)
    df['t_stage_letter'] = df['pathology_t_stage'].str.extract('(\D+)').fillna('')

# Handle eGFR values (converting ">=90" to 90 for analysis)
for col in df.columns:
    if 'egfr' in col and 'value' in col:
        df[col] = df[col].replace('>=90', '90').astype(float)

# Define a function to calculate survival time
def calc_survival_time(row):
    if 'vital_days_after_surgery' in row:
        return row['vital_days_after_surgery']
    else:
        return None

df['survival_time'] = df.apply(calc_survival_time, axis=1)
df['event'] = df['vital_status'].apply(lambda x: 0 if x == 'censored' else 1)

print("Data loaded and preprocessed successfully.")

# ============================================================================
# 1. Demographic Distribution Analysis
# ============================================================================

# Create a function for saving figures
def save_figure(fig, filename):
    fig.savefig(os.path.join(output_dir, filename), dpi=300, bbox_inches='tight')
    plt.close(fig)

print("\n1. Performing Demographic Distribution Analysis...")

# Age distribution by gender
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='age_at_nephrectomy', hue='gender', multiple='stack', bins=20)
plt.title('Age Distribution by Gender', fontsize=14)
plt.xlabel('Age at Nephrectomy', fontsize=12)
plt.ylabel('Count', fontsize=12)
save_figure(plt.gcf(), '1.1_age_distribution_by_gender.png')

# Gender distribution pie chart
plt.figure(figsize=(8, 8))
df['gender'].value_counts().plot.pie(autopct='%1.1f%%', textprops={'fontsize': 12})
plt.title('Gender Distribution', fontsize=14)
plt.ylabel('')
save_figure(plt.gcf(), '1.2_gender_distribution.png')

# BMI correlation with tumor size
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='body_mass_index', y='radiographic_size', hue='malignant')
plt.title('BMI vs Radiographic Tumor Size', fontsize=14)
plt.xlabel('Body Mass Index (BMI)', fontsize=12)
plt.ylabel('Radiographic Size (cm)', fontsize=12)
save_figure(plt.gcf(), '1.3_bmi_vs_tumor_size.png')

# Calculate correlation coefficient
corr = df[['body_mass_index', 'radiographic_size']].corr().iloc[0, 1]
print(f"Correlation between BMI and radiographic size: {corr:.3f}")

# BMI distribution by malignancy
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='malignant', y='body_mass_index')
plt.title('BMI Distribution by Malignancy', fontsize=14)
plt.xlabel('Malignant', fontsize=12)
plt.ylabel('Body Mass Index (BMI)', fontsize=12)
save_figure(plt.gcf(), '1.4_bmi_by_malignancy.png')

# Comorbidity prevalence
comorbidity_cols = [col for col in df.columns if 'comorbidities_' in col]
if comorbidity_cols:
    comorbidity_data = df[comorbidity_cols].sum().sort_values(ascending=False)

    plt.figure(figsize=(14, 8))
    sns.barplot(x=comorbidity_data.values, y=comorbidity_data.index)
    plt.title('Prevalence of Comorbidities', fontsize=14)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Comorbidity', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    save_figure(plt.gcf(), '1.5_comorbidity_prevalence.png')

# Calculate comorbidity prevalence by percentage
comorbidity_percentages = (df[comorbidity_cols].sum() / len(df) * 100).sort_values(ascending=False)
print("\nComorbidity Prevalence (%):")
for comorb, pct in comorbidity_percentages.items():
    clean_name = comorb.replace('comorbidities_', '')
    print(f"{clean_name}: {pct:.1f}%")

# ============================================================================
# 2. Tumor Characteristic Analysis
# ============================================================================

print("\n2. Performing Tumor Characteristic Analysis...")

# Distribution of histologic subtypes
plt.figure(figsize=(12, 6))
subtype_counts = df['tumor_histologic_subtype'].value_counts()
sns.barplot(x=subtype_counts.values, y=subtype_counts.index)
plt.title('Distribution of Tumor Histologic Subtypes', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Tumor Histologic Subtype', fontsize=12)
save_figure(plt.gcf(), '2.1_histologic_subtype_distribution.png')

# Correlation between radiographic and pathologic size
if 'radiographic_size' in df.columns and 'pathologic_size' in df.columns:
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=df, x='radiographic_size', y='pathologic_size')
    plt.plot([0, df[['radiographic_size', 'pathologic_size']].max().max()],
             [0, df[['radiographic_size', 'pathologic_size']].max().max()],
             'r--', label='Perfect agreement')
    plt.title('Radiographic Size vs Pathologic Size', fontsize=14)
    plt.xlabel('Radiographic Size (cm)', fontsize=12)
    plt.ylabel('Pathologic Size (cm)', fontsize=12)
    plt.legend()
    save_figure(plt.gcf(), '2.2_radiographic_vs_pathologic_size.png')

    # Calculate correlation coefficient and mean difference
    rad_vs_path_corr = df[['radiographic_size', 'pathologic_size']].corr().iloc[0, 1]
    rad_vs_path_diff = df['radiographic_size'] - df['pathologic_size']

    print(f"\nRadiographic vs Pathologic Size:")
    print(f"Correlation: {rad_vs_path_corr:.3f}")
    print(f"Mean difference (Radiographic - Pathologic): {rad_vs_path_diff.mean():.3f} cm")
    print(f"Std. deviation of difference: {rad_vs_path_diff.std():.3f} cm")

# ISUP grade distribution by tumor type
if 'tumor_isup_grade' in df.columns and 'tumor_histologic_subtype' in df.columns:
    plt.figure(figsize=(14, 8))
    isup_crosstab = pd.crosstab(df['tumor_histologic_subtype'], df['tumor_isup_grade'])
    isup_crosstab_pct = isup_crosstab.div(isup_crosstab.sum(axis=1), axis=0) * 100

    isup_crosstab_pct.plot(kind='bar', stacked=True)
    plt.title('ISUP Grade Distribution by Tumor Type', fontsize=14)
    plt.xlabel('Tumor Histologic Subtype', fontsize=12)
    plt.ylabel('Percentage', fontsize=12)
    plt.legend(title='ISUP Grade')
    plt.xticks(rotation=45, ha='right')
    save_figure(plt.gcf(), '2.3_isup_grade_by_tumor_type.png')

# TNM stage distribution
t_stage_cols = [col for col in df.columns if 'pathology_t_stage' in col]
if t_stage_cols:
    plt.figure(figsize=(12, 6))
    t_stage_counts = df['pathology_t_stage'].value_counts().sort_index()
    sns.barplot(x=t_stage_counts.index, y=t_stage_counts.values)
    plt.title('Distribution of Pathologic T Stage', fontsize=14)
    plt.xlabel('T Stage', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    save_figure(plt.gcf(), '2.4_t_stage_distribution.png')

# Create TNM combined stage
if 'pathology_t_stage' in df.columns and 'pathology_n_stage' in df.columns and 'pathology_m_stage' in df.columns:
    df['combined_tnm'] = df.apply(lambda row: f"T{row['pathology_t_stage']}N{row['pathology_n_stage']}M{row['pathology_m_stage']}", axis=1)

    plt.figure(figsize=(14, 6))
    combined_tnm_counts = df['combined_tnm'].value_counts().sort_index()
    sns.barplot(x=combined_tnm_counts.index, y=combined_tnm_counts.values)
    plt.title('Combined TNM Stage Distribution', fontsize=14)
    plt.xlabel('TNM Stage', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=90)
    save_figure(plt.gcf(), '2.5_combined_tnm_distribution.png')

# ============================================================================
# 3. Imaging Feature Analysis
# ============================================================================

print("\n3. Performing Imaging Feature Analysis...")

# Function to calculate tumor volume from segmentation masks
def calculate_volumes(case_id):
    case_str = f'case_{case_id:05d}'
    segmentation_path = os.path.join(resampled_dir, case_str, 'segmentation_resampled.nii.gz')

    if not os.path.exists(segmentation_path):
        return None, None

    # Load segmentation mask
    seg_img = sitk.ReadImage(segmentation_path)
    seg_arr = sitk.GetArrayFromImage(seg_img)

    # Get voxel volume in mm
    spacing = seg_img.GetSpacing()
    voxel_volume = spacing[0] * spacing[1] * spacing[2]

    # Calculate volumes (in cm)
    kidney_volume = (seg_arr == 1).sum() * voxel_volume / 1000  # Convert to cm
    tumor_volume = (seg_arr == 2).sum() * voxel_volume / 1000   # Convert to cm

    return kidney_volume, tumor_volume

# Calculate volumes for a subset of cases (for demonstration)
# Adjust the range for your actual analysis
volumes = {}
for case_id in range(min(20, len(df))):  # Limit to 20 cases for example
    try:
        kidney_vol, tumor_vol = calculate_volumes(case_id)
        if kidney_vol is not None:
            case_str = f'case_{case_id:05d}'
            volumes[case_str] = {'kidney_volume': kidney_vol, 'tumor_volume': tumor_vol}
    except Exception as e:
        print(f"Error processing case {case_id}: {e}")

# Convert volumes to DataFrame
volumes_df = pd.DataFrame.from_dict(volumes, orient='index')
volumes_df.index.name = 'case_id'

# Merge volumes with metadata
df_with_volumes = df.join(volumes_df, how='inner')

print(f"Volume analysis completed for {len(volumes_df)} cases")

# Plot distribution of tumor volumes
if not volumes_df.empty:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=volumes_df, x='tumor_volume', bins=15)
    plt.title('Distribution of Tumor Volumes', fontsize=14)
    plt.xlabel('Tumor Volume (cm)', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    save_figure(plt.gcf(), '3.1_tumor_volume_distribution.png')

    # Plot tumor volume vs radiographic size
    if not df_with_volumes.empty:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df_with_volumes, x='radiographic_size', y='tumor_volume', hue='malignant')
        plt.title('Tumor Volume vs Radiographic Size', fontsize=14)
        plt.xlabel('Radiographic Size (cm)', fontsize=12)
        plt.ylabel('Tumor Volume (cm)', fontsize=12)
        save_figure(plt.gcf(), '3.2_tumor_volume_vs_radiographic_size.png')

        # Calculate correlation
        vol_size_corr = df_with_volumes[['tumor_volume', 'radiographic_size']].corr().iloc[0, 1]
        print(f"\nCorrelation between tumor volume and radiographic size: {vol_size_corr:.3f}")

    # Kidney-to-tumor volume ratio
    volumes_df['kidney_to_tumor_ratio'] = volumes_df['kidney_volume'] / volumes_df['tumor_volume']

    plt.figure(figsize=(10, 6))
    sns.histplot(data=volumes_df, x='kidney_to_tumor_ratio', bins=15)
    plt.title('Distribution of Kidney-to-Tumor Volume Ratio', fontsize=14)
    plt.xlabel('Kidney-to-Tumor Volume Ratio', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    save_figure(plt.gcf(), '3.3_kidney_to_tumor_ratio.png')

# Function to calculate tumor heterogeneity (simplified version)
def calculate_heterogeneity(case_id):
    case_str = f'case_{case_id:05d}'
    imaging_path = os.path.join(resampled_dir, case_str, 'imaging_resampled.nii.gz')
    segmentation_path = os.path.join(resampled_dir, case_str, 'segmentation_resampled.nii.gz')

    if not os.path.exists(imaging_path) or not os.path.exists(segmentation_path):
        return None

    # Load imaging and segmentation data
    img = sitk.ReadImage(imaging_path)
    seg = sitk.ReadImage(segmentation_path)

    # Get arrays
    img_arr = sitk.GetArrayFromImage(img)
    seg_arr = sitk.GetArrayFromImage(seg)

    # Extract tumor region (label 2)
    tumor_mask = (seg_arr == 2)
    if not tumor_mask.any():
        return None

    tumor_values = img_arr[tumor_mask]

    # Calculate heterogeneity metrics
    metrics = {
        'mean': np.mean(tumor_values),
        'std': np.std(tumor_values),
        'min': np.min(tumor_values),
        'max': np.max(tumor_values),
        'range': np.max(tumor_values) - np.min(tumor_values),
        'entropy': stats.entropy(np.histogram(tumor_values, bins=20)[0])
    }

    return metrics

# Calculate heterogeneity for a subset of cases
heterogeneity = {}
for case_id in range(min(20, len(df))):  # Limit to 20 cases for example
    try:
        metrics = calculate_heterogeneity(case_id)
        if metrics is not None:
            case_str = f'case_{case_id:05d}'
            heterogeneity[case_str] = metrics
    except Exception as e:
        print(f"Error processing heterogeneity for case {case_id}: {e}")

# Convert heterogeneity to DataFrame
heterogeneity_df = pd.DataFrame.from_dict(heterogeneity, orient='index')
heterogeneity_df.index.name = 'case_id'

print(f"Heterogeneity analysis completed for {len(heterogeneity_df)} cases")

# Plot heterogeneity metrics
if not heterogeneity_df.empty:
    plt.figure(figsize=(12, 8))
    sns.heatmap(heterogeneity_df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Between Tumor Heterogeneity Metrics', fontsize=14)
    save_figure(plt.gcf(), '3.4_heterogeneity_correlation.png')

    # Boxplot of entropy by tumor type
    if not df.empty:
        het_with_metadata = heterogeneity_df.join(df[['tumor_histologic_subtype', 'malignant']], how='inner')

        plt.figure(figsize=(12, 6))
        sns.boxplot(data=het_with_metadata, x='tumor_histologic_subtype', y='entropy')
        plt.title('Tumor Heterogeneity (Entropy) by Histologic Subtype', fontsize=14)
        plt.xlabel('Tumor Histologic Subtype', fontsize=12)
        plt.ylabel('Entropy', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        save_figure(plt.gcf(), '3.5_entropy_by_tumor_type.png')

# ============================================================================
# 4. Outcome Analysis
# ============================================================================

print("\n4. Performing Outcome Analysis...")

# Survival analysis based on tumor characteristics
if 'survival_time' in df.columns and 'event' in df.columns:
    # KM estimator for overall survival
    plt.figure(figsize=(10, 6))
    kmf = KaplanMeierFitter()
    kmf.fit(df['survival_time'], df['event'], label='Overall')
    kmf.plot()
    plt.title('Overall Survival Curve', fontsize=14)
    plt.xlabel('Days After Surgery', fontsize=12)
    plt.ylabel('Survival Probability', fontsize=12)
    save_figure(plt.gcf(), '4.1_overall_survival.png')

    # KM estimator stratified by malignancy
    plt.figure(figsize=(10, 6))
    for malignant, group in df.groupby('malignant'):
        label = 'Malignant' if malignant else 'Benign'
        kmf = KaplanMeierFitter()
        kmf.fit(group['survival_time'], group['event'], label=label)
        kmf.plot()

    plt.title('Survival by Malignancy Status', fontsize=14)
    plt.xlabel('Days After Surgery', fontsize=12)
    plt.ylabel('Survival Probability', fontsize=12)
    save_figure(plt.gcf(), '4.2_survival_by_malignancy.png')

    # Log-rank test for malignancy status
    malignant_group = df[df['malignant'] == True]
    benign_group = df[df['malignant'] == False]

    if not malignant_group.empty and not benign_group.empty:
        results = logrank_test(malignant_group['survival_time'], benign_group['survival_time'],
                              malignant_group['event'], benign_group['event'])
        print(f"\nLog-rank test for survival difference by malignancy:")
        print(f"p-value: {results.p_value:.4f}")

    # Cox Proportional Hazards model
    if len(df) > 10:  # Need enough samples for Cox model
        cph = CoxPHFitter()
        # Select relevant columns for Cox model
        cox_columns = ['age_at_nephrectomy', 'malignant', 'radiographic_size', 'survival_time', 'event']
        cox_df = df[cox_columns].dropna()

        if len(cox_df) > 5:  # Need enough complete cases
            cph.fit(cox_df, duration_col='survival_time', event_col='event')

            plt.figure(figsize=(10, 6))
            cph.plot()
            plt.title('Cox Proportional Hazards Model', fontsize=14)
            save_figure(plt.gcf(), '4.3_cox_model.png')

            print("\nCox Proportional Hazards Model Summary:")
            print(cph.summary)

# eGFR changes pre- and post-operatively
egfr_cols = [col for col in df.columns if 'egfr' in col and 'value' in col]
if len(egfr_cols) >= 2:
    # Extract preop and postop eGFR values
    preop_egfr_col = [col for col in egfr_cols if 'preop' in col][0] if any('preop' in col for col in egfr_cols) else None
    postop_egfr_cols = [col for col in egfr_cols if 'postop' in col]

    if preop_egfr_col and postop_egfr_cols:
        # Plot eGFR change
        plt.figure(figsize=(10, 6))

        # Create a DataFrame for plotting
        egfr_data = []
        for idx, row in df.iterrows():
            if pd.notna(row[preop_egfr_col]):
                egfr_data.append({'case_id': idx, 'timepoint': 'Preop', 'eGFR': row[preop_egfr_col]})

            for col in postop_egfr_cols:
                if pd.notna(row[col]):
                    timepoint = col.replace('_egfr_value', '')
                    egfr_data.append({'case_id': idx, 'timepoint': timepoint, 'eGFR': row[col]})

        egfr_df = pd.DataFrame(egfr_data)

        # Calculate average eGFR by timepoint
        egfr_avg = egfr_df.groupby('timepoint')['eGFR'].mean().reset_index()

        # Plot
        sns.barplot(data=egfr_avg, x='timepoint', y='eGFR')
        plt.title('Average eGFR by Timepoint', fontsize=14)
        plt.xlabel('Timepoint', fontsize=12)
        plt.ylabel('Average eGFR', fontsize=12)
        save_figure(plt.gcf(), '4.4_egfr_by_timepoint.png')

        # Calculate eGFR change
        if 'first_postop_egfr_value' in df.columns and preop_egfr_col in df.columns:
            df['egfr_change'] = df['first_postop_egfr_value'] - df[preop_egfr_col]

            plt.figure(figsize=(10, 6))
            sns.histplot(data=df, x='egfr_change', bins=20)
            plt.axvline(x=0, color='r', linestyle='--')
            plt.title('Distribution of eGFR Change (Postop - Preop)', fontsize=14)
            plt.xlabel('eGFR Change', fontsize=12)
            plt.ylabel('Count', fontsize=12)
            save_figure(plt.gcf(), '4.5_egfr_change_distribution.png')

            print("\neGFR Change Analysis:")
            print(f"Mean eGFR change: {df['egfr_change'].mean():.2f}")
            print(f"Median eGFR change: {df['egfr_change'].median():.2f}")
            print(f"Percentage of patients with decreased eGFR: {(df['egfr_change'] < 0).mean() * 100:.1f}%")

# Predictors of postoperative complications
complication_col = 'clavien_surgical_complications'
if complication_col in df.columns:
    # Convert complications to binary (0 = no complications, 1 = any complication)
    df['has_complication'] = df[complication_col].apply(lambda x: 0 if x == '0' else 1)

    # Risk factors for complications
    potential_risk_factors = ['age_at_nephrectomy', 'body_mass_index', 'radiographic_size', 'malignant']

    plt.figure(figsize=(14, 10))
    for i, factor in enumerate(potential_risk_factors):
        if factor in df.columns:
            plt.subplot(2, 2, i+1)
            sns.boxplot(data=df, x='has_complication', y=factor)
            plt.title(f'{factor} by Complication Status', fontsize=12)
            plt.xlabel('Has Complication', fontsize=10)
            plt.ylabel(factor, fontsize=10)

    plt.tight_layout()
    save_figure(plt.gcf(), '4.6_complication_risk_factors.png')

    # Complication rates by procedure type
    if 'surgical_procedure' in df.columns:
        complication_by_procedure = df.groupby('surgical_procedure')['has_complication'].mean() * 100

        plt.figure(figsize=(12, 6))
        complication_by_procedure.sort_values().plot(kind='barh')
        plt.title('Complication Rate by Surgical Procedure', fontsize=14)
        plt.xlabel('Complication Rate (%)', fontsize=12)
        plt.ylabel('Surgical Procedure', fontsize=12)
        save_figure(plt.gcf(), '4.7_complication_by_procedure.png')

        print("\nComplication Rate by Surgical Procedure:")
        for procedure, rate in complication_by_procedure.sort_values(ascending=False).items():
            print(f"{procedure}: {rate:.1f}%")

# Generate summary report
with open(os.path.join(output_dir, 'analysis_summary.txt'), 'w') as f:
    f.write("KiTS19 Dataset Analysis Summary\n")
    f.write("==============================\n\n")

    f.write("1. Demographic Information\n")
    f.write(f"   - Total cases: {len(df)}\n")
    f.write(f"   - Gender distribution: {df['gender'].value_counts().to_dict()}\n")
    f.write(f"   - Age range: {df['age_at_nephrectomy'].min()}-{df['age_at_nephrectomy'].max()} years (mean: {df['age_at_nephrectomy'].mean():.1f})\n")
    f.write(f"   - BMI mean: {df['body_mass_index'].mean():.2f} (std: {df['body_mass_index'].std():.2f})\n\n")

    f.write("2. Tumor Characteristics\n")
    f.write(f"   - Malignant tumors: {df['malignant'].sum()} ({df['malignant'].mean()*100:.1f}%)\n")
    if 'radiographic_size' in df.columns:
        f.write(f"   - Mean radiographic size: {df['radiographic_size'].mean():.2f} cm\n")
    if 'pathologic_size' in df.columns:
        f.write(f"   - Mean pathologic size: {df['pathologic_size'].mean():.2f} cm\n")
    if 'tumor_histologic_subtype' in df.columns:
        f.write("   - Histologic subtype distribution:\n")
        for subtype, count in df['tumor_histologic_subtype'].value_counts().items():
            f.write(f"     * {subtype}: {count} ({count/len(df)*100:.1f}%)\n")
    f.write("\n")

    f.write("3. Surgical Information\n")
    if 'surgical_procedure' in df.columns:
        f.write("   - Surgical procedure distribution:\n")
        for proc, count in df['surgical_procedure'].value_counts().items():
            f.write(f"     * {proc}: {count} ({count/len(df)*100:.1f}%)\n")
    if 'surgery_type' in df.columns:
        f.write("   - Surgery type distribution:\n")
        for type, count in df['surgery_type'].value_counts().items():
            f.write(f"     * {type}: {count} ({count/len(df)*100:.1f}%)\n")
    if 'operative_time' in df.columns:
        f.write(f"   - Mean operative time: {df['operative_time'].mean():.1f} minutes\n")
    f.write("\n")

    f.write("4. Outcome Analysis\n")
    if 'has_complication' in df.columns:
        f.write(f"   - Overall complication rate: {df['has_complication'].mean()*100:.1f}%\n")
    if 'egfr_change' in df.columns:
        f.write(f"   - Mean eGFR change (if 'egfr_change' in df.columns:
        f.write(f"   - Mean eGFR change (postop - preop): {df['egfr_change'].mean():.2f}\n")
        f.write(f"   - Percentage of patients with decreased eGFR: {(df['egfr_change'] < 0).mean() * 100:.1f}%\n")
    if 'vital_status' in df.columns:
        f.write(f"   - Mortality rate: {(df['vital_status'] != 'censored').mean()*100:.1f}%\n")
    f.write("\n")

    f.write("5. Imaging Analysis\n")
    if not volumes_df.empty:
        f.write(f"   - Mean kidney volume: {volumes_df['kidney_volume'].mean():.2f} cm\n")
        f.write(f"   - Mean tumor volume: {volumes_df['tumor_volume'].mean():.2f} cm\n")
        f.write(f"   - Mean kidney-to-tumor ratio: {volumes_df['kidney_to_tumor_ratio'].mean():.2f}\n")
    if not heterogeneity_df.empty:
        f.write(f"   - Mean tumor heterogeneity (entropy): {heterogeneity_df['entropy'].mean():.2f}\n")
    f.write("\n")

    f.write("6. Key Correlations\n")
    if 'radiographic_size' in df.columns and 'pathologic_size' in df.columns:
        f.write(f"   - Radiographic vs Pathologic size: {df[['radiographic_size', 'pathologic_size']].corr().iloc[0, 1]:.3f}\n")
    if 'body_mass_index' in df.columns and 'radiographic_size' in df.columns:
        f.write(f"   - BMI vs Tumor size: {df[['body_mass_index', 'radiographic_size']].corr().iloc[0, 1]:.3f}\n")
    if not df_with_volumes.empty and 'tumor_volume' in df_with_volumes.columns and 'radiographic_size' in df_with_volumes.columns:
        f.write(f"   - Tumor volume vs Radiographic size: {df_with_volumes[['tumor_volume', 'radiographic_size']].corr().iloc[0, 1]:.3f}\n")

print("Analysis complete! Results saved to", output_dir)

# ============================================================================
# Additional Analysis: Machine Learning Prediction Models
# ============================================================================

print("\n5. Building Prediction Models...")

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
from sklearn.impute import SimpleImputer

# 1. Malignancy Prediction Model
if 'malignant' in df.columns:
    # Select features for prediction
    features = ['age_at_nephrectomy', 'gender', 'body_mass_index', 'radiographic_size']

    # Convert gender to numeric
    X_data = df[features].copy()
    if 'gender' in X_data.columns:
        X_data['gender'] = X_data['gender'].map({'male': 1, 'female': 0})

    # Target variable
    y_data = df['malignant'].astype(int)

    # Remove rows with NaN
    valid_idx = ~X_data.isna().any(axis=1)
    X_data = X_data[valid_idx]
    y_data = y_data[valid_idx]

    if len(X_data) > 20:  # Need enough samples for modeling
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3, random_state=42)

        # Create and train model
        model = Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler()),
            ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
        ])

        model.fit(X_train, y_train)

        # Evaluate model
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        print("\nMalignancy Prediction Model Performance:")
        print(classification_report(y_test, y_pred))

        # ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        auc = roc_auc_score(y_test, y_prob)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for Malignancy Prediction')
        plt.legend()
        save_figure(plt.gcf(), '5.1_malignancy_prediction_roc.png')

        # Feature importance
        if hasattr(model['classifier'], 'feature_importances_'):
            importances = model['classifier'].feature_importances_
            feature_importance = pd.Series(importances, index=X_train.columns)

            plt.figure(figsize=(10, 6))
            feature_importance.sort_values().plot(kind='barh')
            plt.title('Feature Importance for Malignancy Prediction')
            plt.xlabel('Importance')
            save_figure(plt.gcf(), '5.2_malignancy_feature_importance.png')

# 2. Postoperative Complication Prediction
if 'has_complication' in df.columns:
    # Select features for complication prediction
    complication_features = ['age_at_nephrectomy', 'body_mass_index', 'radiographic_size',
                            'malignant', 'operative_time']

    # Filter columns that exist in the dataframe
    available_features = [f for f in complication_features if f in df.columns]

    if len(available_features) >= 3:  # Need at least a few features
        X_comp = df[available_features].copy()
        y_comp = df['has_complication'].astype(int)

        # Remove rows with NaN
        valid_idx = ~X_comp.isna().any(axis=1)
        X_comp = X_comp[valid_idx]
        y_comp = y_comp[valid_idx]

        if len(X_comp) > 20:  # Need enough samples
            # Split data
            X_train, X_test, y_train, y_test = train_test_split(X_comp, y_comp, test_size=0.3, random_state=42)

            # Create and train model
            comp_model = Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler()),
                ('classifier', LogisticRegression(max_iter=1000, random_state=42))
            ])

            comp_model.fit(X_train, y_train)

            # Evaluate model
            y_pred = comp_model.predict(X_test)
            y_prob = comp_model.predict_proba(X_test)[:, 1]

            print("\nComplication Prediction Model Performance:")
            print(classification_report(y_test, y_pred))

            # ROC curve
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            auc = roc_auc_score(y_test, y_prob)

            plt.figure(figsize=(8, 6))
            plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
            plt.plot([0, 1], [0, 1], 'k--')
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('ROC Curve for Complication Prediction')
            plt.legend()
            save_figure(plt.gcf(), '5.3_complication_prediction_roc.png')

            # Model coefficients (for logistic regression)
            if hasattr(comp_model['classifier'], 'coef_'):
                coefs = comp_model['classifier'].coef_[0]
                coef_df = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': coefs})
                coef_df = coef_df.sort_values('Coefficient', ascending=False)

                plt.figure(figsize=(10, 6))
                sns.barplot(data=coef_df, x='Coefficient', y='Feature')
                plt.title('Feature Coefficients for Complication Prediction')
                plt.axvline(x=0, color='k', linestyle='--')
                save_figure(plt.gcf(), '5.4_complication_coefficients.png')

# ============================================================================
# Create integrated visualization
# ============================================================================

# Function to create an integrated visualization for a specific case
def create_integrated_case_analysis(case_id):
    case_str = f'case_{case_id:05d}'
    if case_str not in df.index:
        print(f"Case {case_str} not found in metadata")
        return

    try:
        # Get case metadata
        case_meta = df.loc[case_str]

        # Check if volume data is available
        case_volumes = volumes_df.loc[case_str] if case_str in volumes_df.index else None

        # Load image and segmentation
        volume, segmentation = load_resampled_case(case_id)
        volume_data = volume.get_fdata()
        segmentation_data = segmentation.get_fdata()

        # Create figure with multiple panels
        fig = plt.figure(figsize=(15, 10))

        # Panel 1: Axial slice with segmentation overlay
        ax1 = plt.subplot2grid((2, 3), (0, 0))
        slice_idx_axial = volume_data.shape[2] // 2
        axial_slice = volume_data[:, :, slice_idx_axial]
        axial_seg = segmentation_data[:, :, slice_idx_axial]

        # Normalize and convert to RGB
        axial_slice_norm = np.clip(axial_slice, -512, 512)
        axial_slice_norm = (axial_slice_norm + 512) / 1024 * 255
        axial_rgb = np.stack([axial_slice_norm]*3, axis=-1).astype(np.uint8)

        # Create colored segmentation mask
        seg_colored = np.zeros_like(axial_rgb)
        seg_colored[axial_seg == 1] = [255, 0, 0]  # Kidney in red
        seg_colored[axial_seg == 2] = [0, 0, 255]  # Tumor in blue

        # Overlay segmentation
        alpha = 0.5
        segbin = axial_seg > 0
        axial_overlayed = np.where(
            segbin[..., None],
            np.round(alpha * seg_colored + (1 - alpha) * axial_rgb).astype(np.uint8),
            axial_rgb
        )

        ax1.imshow(axial_overlayed)
        ax1.set_title(f'Axial Slice (Red=Kidney, Blue=Tumor)', fontsize=10)
        ax1.axis('off')

        # Panel 2: Key metadata
        ax2 = plt.subplot2grid((2, 3), (0, 1), colspan=2)
        key_fields = [
            f"Age: {case_meta['age_at_nephrectomy']} years",
            f"Gender: {case_meta['gender']}",
            f"BMI: {case_meta['body_mass_index']:.1f}",
            f"Malignant: {case_meta['malignant']}",
            f"Histology: {case_meta['tumor_histologic_subtype'] if 'tumor_histologic_subtype' in case_meta else 'Unknown'}",
            f"Radiographic Size: {case_meta['radiographic_size']:.1f} cm" if 'radiographic_size' in case_meta else "Size: Unknown",
            f"Pathologic Size: {case_meta['pathologic_size']:.1f} cm" if 'pathologic_size' in case_meta else "",
            f"Surgery: {case_meta['surgical_procedure'] if 'surgical_procedure' in case_meta else 'Unknown'}",
            f"T-Stage: {case_meta['pathology_t_stage'] if 'pathology_t_stage' in case_meta else 'Unknown'}"
        ]

        if case_volumes is not None:
            key_fields.extend([
                f"Kidney Volume: {case_volumes['kidney_volume']:.1f} cm",
                f"Tumor Volume: {case_volumes['tumor_volume']:.1f} cm",
                f"Kidney-Tumor Ratio: {case_volumes['kidney_to_tumor_ratio']:.1f}"
            ])

        ax2.text(0.05, 0.95, "\n".join(key_fields), transform=ax2.transAxes,
                fontsize=12, va='top')
        ax2.axis('off')

        # Panel 3: Sagittal view
        ax3 = plt.subplot2grid((2, 3), (1, 0))
        slice_idx_sagittal = volume_data.shape[0] // 2
        sagittal_slice = volume_data[slice_idx_sagittal, :, :]
        sagittal_seg = segmentation_data[slice_idx_sagittal, :, :]

        # Normalize and process similarly to axial
        sagittal_slice_norm = np.clip(sagittal_slice, -512, 512)
        sagittal_slice_norm = (sagittal_slice_norm + 512) / 1024 * 255
        sagittal_rgb = np.stack([sagittal_slice_norm]*3, axis=-1).astype(np.uint8)

        # Create colored segmentation mask
        seg_colored = np.zeros_like(sagittal_rgb)
        seg_colored[sagittal_seg == 1] = [255, 0, 0]  # Kidney in red
        seg_colored[sagittal_seg == 2] = [0, 0, 255]  # Tumor in blue

        # Overlay segmentation
        segbin = sagittal_seg > 0
        sagittal_overlayed = np.where(
            segbin[..., None],
            np.round(alpha * seg_colored + (1 - alpha) * sagittal_rgb).astype(np.uint8),
            sagittal_rgb
        )

        ax3.imshow(sagittal_overlayed)
        ax3.set_title('Sagittal View', fontsize=10)
        ax3.axis('off')

        # Panel 4: Coronal view
        ax4 = plt.subplot2grid((2, 3), (1, 1))
        slice_idx_coronal = volume_data.shape[1] // 2
        coronal_slice = volume_data[:, slice_idx_coronal, :]
        coronal_seg = segmentation_data[:, slice_idx_coronal, :]

        # Normalize and process similarly to axial
        coronal_slice_norm = np.clip(coronal_slice, -512, 512)
        coronal_slice_norm = (coronal_slice_norm + 512) / 1024 * 255
        coronal_rgb = np.stack([coronal_slice_norm]*3, axis=-1).astype(np.uint8)

        # Create colored segmentation mask
        seg_colored = np.zeros_like(coronal_rgb)
        seg_colored[coronal_seg == 1] = [255, 0, 0]  # Kidney in red
        seg_colored[coronal_seg == 2] = [0, 0, 255]  # Tumor in blue

        # Overlay segmentation
        segbin = coronal_seg > 0
        coronal_overlayed = np.where(
            segbin[..., None],
            np.round(alpha * seg_colored + (1 - alpha) * coronal_rgb).astype(np.uint8),
            coronal_rgb
        )

        ax4.imshow(coronal_overlayed)
        ax4.set_title('Coronal View', fontsize=10)
        ax4.axis('off')

        # Panel 5: 3D rendering placeholder or other analysis
        ax5 = plt.subplot2grid((2, 3), (1, 2))

        # Create a simple histogram of the tumor region
        tumor_mask = segmentation_data == 2
        if tumor_mask.any():
            tumor_values = volume_data[tumor_mask]
            ax5.hist(tumor_values, bins=50, color='blue', alpha=0.7)
            ax5.set_title('Tumor Intensity Histogram', fontsize=10)
            ax5.set_xlabel('Hounsfield Units (HU)')
            ax5.set_ylabel('Frequency')
        else:
            ax5.text(0.5, 0.5, "No tumor detected", ha='center', va='center')
            ax5.axis('off')

        plt.tight_layout()
        plt.suptitle(f'Integrated Analysis for Case {case_str}', fontsize=16, y=1.02)

        # Save figure
        integrated_dir = os.path.join(output_dir, 'integrated_analysis')
        os.makedirs(integrated_dir, exist_ok=True)
        plt.savefig(os.path.join(integrated_dir, f'{case_str}_integrated.png'), dpi=200, bbox_inches='tight')
        plt.close()

        print(f"Integrated analysis created for case {case_str}")
        return True

    except Exception as e:
        print(f"Error creating integrated analysis for case {case_id}: {e}")
        return False

# Create integrated visualizations for a few example cases
successful = 0
for case_id in range(min(5, len(df))):
    if create_integrated_case_analysis(case_id):
        successful += 1
    if successful >= 3:  # Limit to 3 successful examples
        break

print(f"\nAnalysis complete! All results saved to {output_dir}")
print("Generated:")
print("- Demographic analysis charts")
print("- Tumor characteristic analysis")
print("- Volume and heterogeneity analysis")
print("- Survival and outcome analysis")
print("- Prediction models")
print("- Integrated case visualizations")
```
